```python
import csv
import pandas as pd
```

```python
models = ['qanet-01', 'qanet-03', 'qanet-05', 'uqanet-01', 'uqanet-03', 'uqanet-05', 'uqanet-11']

w = [0.96,0.98,1,0.99,0.97,0.95,0.94]

model_weight = dict(zip(models, w))
model_weight

anw_len_weight = [0.5245098039215687,
                 0.16700473292765383,
                 0.10378634212305612,
                 0.07606490872210954,
                 0.04006085192697769,
                 0.02636916835699797,
                 0.016396213657876944,
                 0.013015551048005409,
                 0.007944557133198107,
                 0.006254225828262339,
                 0.004394861392832995,
                 0.0042258282623394185,
                 0.003211629479377958,
                 0.002704530087897228,
                 0.002535496957403651,
                 0.0015212981744421906]
```

```python
split = 'dev'
```

```python
res = {}
for model in models:
    with open(f'save/test/{model}/{split}_submission.csv', encoding='utf-8') as csvfile:
        csv_reader = csv.reader(csvfile)
        tmp = []
        for row in csv_reader:
            tmp.append(row)
    res[model] = tmp
```

```python
len_count = [0]*16
for i in range(1, len(res[models[0]])):
    for model in models:
        model_pred = res[model][i] 
        uuid, answer = model_pred
        len_count[len(answer.split())] += 1
```

```python
tmp = pd.Series(len_count)
print(tmp)
tmp = tmp/sum(tmp)
anw_len_weight = pd.Series(anw_len_weight)/tmp
```

```python
with open(f'ensemble_{split}_submission.csv', 'w', newline='', encoding='utf-8') as csv_fh:
    csv_writer = csv.writer(csv_fh, delimiter=',')
    csv_writer.writerow(['Id', 'Predicted'])
    for i in range(1, len(res[models[0]])):
        res_weight = {}
        for model in models:
            model_pred = res[model][i] 
            uuid, answer = model_pred
            if answer in res_weight:
                res_weight[answer] += model_weight[model]
            else:
                res_weight[answer] = model_weight[model]
        
        res_weight = {r: w*anw_len_weight[len(r.split())] for r, w in res_weight.items()}
        weight_res = [(y, x) for x, y in res_weight.items()]
        weight_res.sort(reverse=True)
        print(weight_res)
        final_answer = weight_res[0][1]
        csv_writer.writerow([uuid, final_answer])
```
